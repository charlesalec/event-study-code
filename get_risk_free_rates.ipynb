```python
import requests
import pandas as pd

# Alpha Vantage API key
api_key = '56TS94GBJS6B8P7X'

# URL to request U.S. Treasury Yield data (example for the 10-year yield)
url = f'https://www.alphavantage.co/query?function=TREASURY_YIELD&interval=daily&maturity=10year&apikey={api_key}'

# Request the data
response = requests.get(url)
data = response.json()

# Extract the 'data' part which contains daily yields
yield_data = data['data']

# Convert the data into a pandas DataFrame
df = pd.DataFrame(yield_data)

# Convert the 'date' column to datetime
df['date'] = pd.to_datetime(df['date'])

# Set the 'date' column as the index for easier filtering
df.set_index('date', inplace=True)

# Convert the 'value' column to numeric type
df['value'] = pd.to_numeric(df['value'], errors='coerce')

print(df['value'])

# Specify the date you want to query (example: '2022-01-05')
query_date = '2022-01-05'

# Get the risk-free rate (Treasury yield) for the specific date
risk_free_rate = df.loc[query_date]['value']

# Display the result
print(f"The 10-year Treasury yield on {query_date} is {risk_free_rate}%")
```

```python
def annual_to_daily_yield(annual_yield):
    """
    Convert annual yield to daily yield assuming 260 trading days in a year.
    
    Parameters:
    annual_yield (float): The annual yield in percentage (e.g., 1.71 for 1.71%)
    
    Returns:
    float: The daily compounded yield in percentage
    """
    daily_yield = ((1 + annual_yield / 100) ** (1 / 260) - 1) * 100
    return daily_yield

# Example annual yield of 1.71%
annual_yield = risk_free_rate
daily_yield = annual_to_daily_yield(annual_yield)

print(f"Annual Yield: {annual_yield}%")
print(f"Daily Yield: {daily_yield:.5f}%")
```

```python
daily_yields = []
for rf in df['value']:
    daily_yields.append(annual_to_daily_yield(rf))
    df.loc[query_date, 'daily_value'] = annual_to_daily_yield(rf)

print(daily_yields)

# Specify the date you want to query (example: '2022-01-05')
query_date = '2022-01-05'

# Get the risk-free rate (Treasury yield) for the specific date
risk_free_rate = df.loc[query_date]['daily_value']

# Display the result
print(f"The daily 10-year Treasury yield on {query_date} is {risk_free_rate}%")
```
